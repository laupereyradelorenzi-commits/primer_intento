```python
import numpy as np

data = np.array([1,0])
print(data[0])
print(type(data[0]))

data = np.array([[1,0],[2,0],[3,0]])
print(data[0][:0])
```

```python
print(data.shape)
print(data.size)
print(data.ndim)
print(data.dtype)
```

```python
data = np.array([[1,0],[2,0]],dtype=np.int32)
data = np.array([[1,0],[2,0]],dtype=np.complex64)
data = np.array([[1,0],[2,0]],dtype=np.float16)
# Conversi√≥n
data = np.array(data,dtype=np.uint)
data = data.astype(np.bool_)
print(data)
```

```python
data = np.array(range(10))
print(data)

data = np.arange(10)
print(data)

data = np.zeros((10,5))
# https://numpy.org/doc/stable/reference/generated/numpy.zeros.html
print(data)

data = np.ones(10)
print(data)
```

```python
shape = (10,80) # 10 rows x 80 cols
data = np.zeros(shape)
print(data)
```

```python
data = np.linspace(0,1,10)
print(data)
```

```python
data = np.logspace(0,2,10) # 10puntos entre 2**0 y 2**2
print(data)
```

```python
data = np.identity(3)
print(data)
print("-"*10)

data = np.eye(3,k=1) # https://numpy.org/doc/stable/reference/generated/numpy.eye.html
print(data)
print("-"*10)

data = np.diag(range(1,4))
print(data)
```

```python
data =np.array(range(1,4))
repeated_data = np.repeat(data, 3)
resultado = np.concatenate((repeated_data, data))
print(resultado)
```
